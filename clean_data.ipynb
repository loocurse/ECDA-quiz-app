```python
import pandas as pd
import json
import copy
```

```python
df = pd.read_csv('Prototype content.csv', encoding='ISO-8859-1')

df
```

```python
questions = df.loc[df['STUDENT ID'] == 'S0001'][['DOMAIN', 'SUB CATEGORY', 'SKILL']]
questions.drop_duplicates(subset=['SKILL'], inplace=True)
questions
```

```python
output = {}

free_text_questions = [
  {
      'question': 'Commendable Areas',
      'type': 'text_area',
      'response': None
  },
]

for idx, row in questions.iterrows():
  if not output.get(row['DOMAIN']):
    output[row['DOMAIN']] = {}
  if not output[row['DOMAIN']].get(row['SUB CATEGORY']):
    output[row['DOMAIN']][row['SUB CATEGORY']] = copy.copy(free_text_questions)
  output[row['DOMAIN']][row['SUB CATEGORY']].insert(0, {'question': row['SKILL'], 'response': None})


json.dump(output, open('updatedQuestions.json', 'w'), indent=4)
```
